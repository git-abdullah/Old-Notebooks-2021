```python
df = df[["country", "year", "sex", "age", "suicides_no"]]
```

```python
su_count = df[['country', 'suicides_no']]
a = su_count['country'].value_counts()[0:100:5]
plt.figure(figsize=(30, 6))
plt.plot(a, label="Suicides_no")
plt.legend()
```

```python
a[0:100:5]
```

```python
df_year = df[["year", "suicides_no"]]
year_su = df_year.groupby('year').mean()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 6))
plt.plot(year_su, label="%age Suicides")
plt.legend()
```

```python
age_based = df[["age", "suicides_no"]]
age_based_f = df['age'].value_counts()
plt.figure(figsize=(10, 5))
plt.plot(age_based_f, label='Suicides(age)')
plt.legend()
```
