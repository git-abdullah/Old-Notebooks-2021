```python
import pandas as pd
csv = pd.read_csv("AMDPM_18-7-2021_022646.csv")

def get_seconds(date):
    second = date[16:len(date)-3]
    return second

def get_hz(util):
    return (util/100)*3400


csv['seconds'] = csv['Metric Id'].apply(get_seconds)
csv['Core 1 Clock'] = csv['CPU Core1 Utilization'].apply(get_hz)
csv['Core 2 Clock'] = csv['CPU Core2 Utilization'].apply(get_hz)
#df = csv[['seconds', 'Core 1 Clock', 'Core 2 Clock', 'AMD Radeon HD 7480D Speed', 'AMD Radeon HD 7480D Temperature']]

cpu1 = csv['Core 1 Clock'].values
cpu2 = csv['Core 2 Clock'].values
gpu = csv['AMD Radeon HD 7480D Speed'].values
tem = csv['AMD Radeon HD 7480D Temperature'].values

csv
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 7))
plt.plot(cpu1, label="CPU Core1 Clock (Max 3400MHz)")
plt.plot(cpu2, label="CPU Core2 Clock (Max 3400MHz)")
plt.plot(gpu, label="AMD Radeon HD 7480D Clock (Max 723 MHz)")
plt.legend()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(tem, label="AMD Radeon HD 7480D Temperature", color="red")
plt.legend()
```
