```python
import math
import matplotlib.pyplot as plt

def plot_graph(x, y):
    fig, (plt1, plt2) = plt.subplots(1, 2, figsize=(9, 4))
    fig.suptitle("Projectile Motion in 2D")

    plt1.plot(x, y)
    plt1.set_title('Free Scale')
    plt1.set_xlabel('Horizontal Range')
    plt1.set_ylabel('Height')


    plt2.plot(x, y)
    plt2.set_title('Fixed Scale')
    plt2.set_xlabel('Horizontal Range')
    plt2.set_ylabel('Height')
    plt.xlim(-300, 300)
    plt.ylim(0, 200)
    
    plt.show()

def frange(start, stop, interval):
    numbers = []
    while start < stop:
        numbers.append(start)
        start += interval
    return numbers

def projectile_motion(velocity_i, angle_thetta):
    angle_thetta = math.radians(angle_thetta)
    g = 9.8

    time_ofFlight = 2 * velocity_i * math.sin(angle_thetta) / g 
    print(f"Time of flight: {time_ofFlight}")

    H_max = (velocity_i ** 2) * (math.sin(angle_thetta)) ** 2 / (2 * g)
    print(f"Maximum Projectile Height: {H_max}")

    h_range = (velocity_i ** 2) * 2 * (math.sin(angle_thetta) * math.cos(angle_thetta)) / g
    print(f"Range: {h_range}")
   
    x = []
    y = []

    intervals = frange(0, time_ofFlight, 0.001)

    for t in intervals:
        w = velocity_i * math.cos(angle_thetta) * t
        x.append(w)
        w1 = velocity_i * math.sin(angle_thetta) * t - 0.5 * g * t * t
        y.append(w1)
    
    plot_graph(x, y)


projectile_motion(50, 37)
```
